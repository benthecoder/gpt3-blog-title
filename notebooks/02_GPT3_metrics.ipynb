```python
import wandb
api = wandb.Api()
import pandas as pd

runs = [
    ("adav1", "ft-uSwqlpAX9RlIcFfrYkyNI8g1"),
    ("adav2", "ft-2ooqd5t0ycMwvf8TM3MUuyAt"),
    ("adav3", "ft-cNrLBjqB9V6g0WLayqpRxcGK"),
    ("babbage", "ft-3Svmi2GPLKTLnZAQCmHlEYql"),
]

df = pd.DataFrame(
    [
        api.run(f"benneo/GPT-3 blog title/{run_id}").summary_metrics
        for project_name, run_id in runs
    ],
    index=[project_name for project_name, _ in runs],
)


# remove "classification/" from column names
df.columns = [col.split("/")[-1] for col in df.columns]

# filter auprc, aorc, precision, validation token accuarcy, validation sequence accuracy
df = df.filter(regex="auprc|f1.0|auroc|accuracy|precision|recall|token|sequence")
df
```
